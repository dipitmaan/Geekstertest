```python

# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "C:/Users/hp/Downloads/BAC.csv"
df = pd.read_csv(file_path)

# Display the first few rows of the dataset
df.head()

# Check the column names and data types

print("Column names in the dataset:", df.columns)
print("Data types in the dataset:\n", df.dtypes)

# Summary statistics of the dataset
print("Summary statistics:\n", df.describe())

# Line Plot
# A line plot is used to display data points over time. Its useful for visualizing trends in data.

plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Close'], label='Closing Price', color='blue')
plt.title('Line Plot of BAC Stock Closing Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.xticks(rotation=45)
plt.legend()
plt.show()

# Bar Plot
# Creating a bar plot to show the average closing price by month.

# Convert 'Date' to datetime and extract the month
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month

# Group by month and calculate average closing price
monthly_avg = df.groupby('Month')['Close'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='Month', y='Close', data=monthly_avg, palette='viridis')
plt.title('Average Monthly Closing Prices')
plt.xlabel('Month')
plt.ylabel('Average Closing Price')
plt.show()

# Scatter Plot
# A scatter plot shows the relationship between two continuous variables.

plt.figure(figsize=(12, 6))
sns.scatterplot(x='Volume', y='Close', data=df, color='red')
plt.title('Scatter Plot of Volume vs. Closing Price')
plt.xlabel('Volume')
plt.ylabel('Closing Price')
plt.show()

# Histogram
# A histogram is used to represent the distribution of numerical data by dividing it into bins.

plt.figure(figsize=(12, 6))
plt.hist(df['Close'], bins=20, edgecolor='black', color='green')
plt.title('Histogram of BAC Stock Closing Prices')
plt.xlabel('Closing Price')
plt.ylabel('Frequency')
plt.show()

# Box Plot
# A box plot displays the distribution of data based on a five-number summary.

plt.figure(figsize=(12, 6))
sns.boxplot(x='Month', y='Close', data=df, palette='Set2')
plt.title('Box Plot of Closing Prices by Month')
plt.xlabel('Month')
plt.ylabel('Closing Price')
plt.show()

# Heatmap
# A heatmap is a graphical representation of data where individual values are represented by colors.

plt.figure(figsize=(10, 6))
corr_matrix = df[['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap of Correlation Matrix')
plt.show()

# Pair Plot
# A pair plot is used to visualize relationships between multiple variables in a dataset.

sns.pairplot(df[['Open', 'High', 'Low', 'Close', 'Volume']])
plt.suptitle('Pair Plot of Selected Financial Metrics', y=1.02)
plt.show()

# Conclusion
# In this notebook, we explored various types of graphs, including line plots, bar plots, scatter plots, histograms, box plots, heatmaps, and pair plots. Each type of graph was generated using the BAC stock price dataset, and we discussed their characteristics and appropriate use cases. This exploration provides a solid foundation for selecting the right graph type for data visualization tasks.
```
